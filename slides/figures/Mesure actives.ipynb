```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("new_ratio_global.csv", 
                 index_col=["time"])
ax = df.drop("ratio legit", 1).rename(
            columns={"tx route":             "tx route - sans corr.",
                     "tx route calibration": "tx route - avec corr.",
                     "rx route":             "rx route - sans corr.",
                     "rx route calibration": "rx route - avec corr."}).plot()
ax.set_ylabel("$\delta$", rotation=0)
plt.tight_layout()
plt.savefig("mesure_active.pdf")
```
